### Important! As I smallest triangle i define a triangle with a minimal perimeter

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import itertools

start_point = -100
end_point = 100

def distance_beetween(p1,p2) -> float:
    """Function to compute distance between 2 points"""
    return ((p2[0] - p1[0])**2 + (p2[1] - p2[0])**2)**0.5

def is_valid(dist_1, dist_2, dist_3) -> bool:
    """Checking if from 3 points  with given distance can we make a triangle"""

  
    lst = [dist_1, dist_2, dist_3]
    #Computing max section and checking condtion
    arg_max = np.argmax(lst)
    if lst.pop(arg_max) < sum(lst):
        return True
    
    return False
     

def plot_figure(points_lst):
    """Plot points"""
    x_coor = [elem[0] for elem in points_lst]
    y_coor = [elem[1] for elem in points_lst]
    plt.scatter(x_coor, y_coor)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Points on 2d plane')
    plt.xlim([start_point, end_point])
    plt.ylim([start_point, end_point])

def tick_triangle(points):
    x_coor = [elem[0] for elem in points]
    y_coor = [elem[1] for elem in points]
    plt.triplot(x_coor, y_coor, color = 'm')

    

def min_triangle(n):
    "Function to generate data and find smallest triangle"
    if not isinstance(n, int) or n < 0:
        print('Use proper value of points number')
        return None
    
    points_lst = []
    for i in range(n):
        x = np.round(random.uniform(start_point, end_point), 2)
        y = np.round(random.uniform(start_point, end_point), 2)
        points_lst.append((x,y))
        
    min_dist = np.inf
    min_points = [np.NaN, np.NaN, np.NaN]
    comb_list = list(itertools.combinations(points_lst,3))
    for elem in comb_list:

        p1, p2, p3 = elem
        dist_1 = distance_beetween(p1, p2)
        dist_2 = distance_beetween(p2, p3)
        dist_3 = distance_beetween(p1, p3)
        if is_valid(dist_1, dist_2, dist_3):
            if dist_1 + dist_2 + dist_3 < min_dist:
                min_dist = dist_1 + dist_2 + dist_3
                min_points = [p1,p2,p3]
        
    print(f'Points: {min_points}')
    print(f'Minimal perimeter: {min_dist}')
    
    #Comment this 2 lines if do not want to see visualization(or not able)
    plot_figure(points_lst)
    tick_triangle(min_points)

```
